```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import StaleElementReferenceException
import urllib.request
import base64
import numpy as np
import cv2



driver = webdriver.Chrome(ChromeDriverManager().install())
url=driver.get("https://www.wedressfair.fr/femme")
wait = WebDriverWait(driver, 20)
button = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[7]/div/div/div/div/div/div/button")))
button.click()
```

```python
import time

info_list=[]
seen_items = set()
pageNumber=1
while True:
    try:
        
        cards=driver.find_elements("xpath","//div[@class='group relative flex-shrink-0']") 
        for i, card in zip(range(len(cards)), cards):
            dict = {}

            image_src = driver.find_elements("xpath", f'//*[@id="products-container"]/div[{i}]/a/div[1]/img')
            for a in image_src:
                dict["Image_Source"] = a.get_attribute("src")

            marque = driver.find_elements("xpath", f'//*[@id="products-container"]/div[{i}]/a/div[2]/div[1]/p[1]')
            for b in marque:
                dict["Marque"] = b.text
            
            description = driver.find_elements("xpath", f'//*[@id="products-container"]/div[{i}]/a/div[2]/div[1]/p[2]')
            for c in description:
                dict["Description"] = c.text
            
            prix = driver.find_elements("xpath", f'//*[@id="products-container"]/div[{i}]/a/div[2]/div[2]/div/div/span')
            for d in prix:
                dict["Prix"] = d.text
            
            taille = driver.find_elements("xpath", f'//*[@id="products-container"]/div[{i}]/a/div[1]')
            for e in taille:
                dict["Taille"] = e.text
            
            lien = driver.find_elements("xpath", f'//*[@id="products-container"]/div[{i}]/a')
            for f in lien :
                dict["Lien"]= f.get_attribute("href")
            
            info_list.append(dict)
            
            
            driver.execute_script("arguments[0].scrollIntoView();",card)
            time.sleep(0.2)

        WebDriverWait(driver, 30)
        next_button = WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, f'//a[@data-value="{str(pageNumber)}"]')))          
        next_button.click()
        
        pageNumber+=1
        
        
        
        
    except StaleElementReferenceException:
        print("Stale element reference exception occurred. Waiting for the page to load...")
        continue

    except:
        print("Unexpected exception occurred.")
        break
len(info_list)
```

```python
import pandas as pd

data = pd.read_csv('donnée_wedressfair.csv')
data
```

```python
import cv2
import numpy as np
import urllib.request



def ar_image(filepath):
    
    resp = urllib.request.urlopen(filepath)
    image_bytes = np.asarray(bytearray(resp.read()), dtype="uint8")
    img = cv2.imdecode(image_bytes, cv2.IMREAD_COLOR)
    
    return img

data['Array_image'] = data.apply(lambda x: ar_image(x['Image_Source']), axis=1)
```

```python
y = top_10_vetements = list(data['Vetement'].value_counts()[:10].index)
data.loc[~data['Vetement'].isin(top_10_vetements), 'Vetement'] = 'Other'
```

```python
apparel_list = data['Vetement'].unique()
data['label'] = pd.Categorical(data['Vetement'], categories=apparel_list).codes + 1
```

```python
data_ = data.drop('Array_image', axis=1)

data_.to_csv('donnée_wedressfair_2.csv', index=False)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from skimage.transform import resize
from tensorflow import keras

X_train, X_test, y_train, y_test = train_test_split(data['Array_image'], data['label'], test_size=0.2, random_state=7)

x_train, _, _, _ = train_test_split(data['Array_image'], data['label'], test_size=0.2, random_state=7)


desired_shape = (100, 100, 3)

resized_images = []
for image in x_train:
    resized_image = resize(image, desired_shape)
    resized_images.append(resized_image)

x_train = np.stack(resized_images, axis=0)

x_train = x_train.reshape((x_train.shape[0], desired_shape[0], desired_shape[1], desired_shape[2]))

x_train = x_train.astype('float32') / 255.0

num_classes = 12  
y_train = keras.utils.to_categorical(y_train, num_classes)

model = keras.models.Sequential()

model.add(keras.layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(desired_shape[0], desired_shape[1], desired_shape[2])))
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
model.add(keras.layers.Dropout(0.25))





model.add(keras.layers.Flatten(input_shape=(desired_shape[0], desired_shape[1], desired_shape[2])))
model.add(keras.layers.Dense(num_classes, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100, batch_size=32)

```

```python
model.save('model.h5')
```

```python
from tensorflow.keras.models import load_model

model = load_model('model.h5')
```

```python

resized_test_images = []
for image in X_test:
    resized_image = resize(image, desired_shape)
    resized_test_images.append(resized_image)

x_test_processed = np.stack(resized_test_images, axis=0)
x_test_processed = x_test_processed.astype('float32') / 255.0

x_test_processed = x_test_processed[:y_test.shape[0]]

y_test_encoded = keras.utils.to_categorical(y_test, num_classes=12)

test_loss, test_accuracy = model.evaluate(x_test_processed, y_test_encoded)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
def teste ( image_s ):

    image = ar_image(image_s)
    
    resized_image = resize(image, desired_shape)
    x_new_processed = np.expand_dims(resized_image, axis=0)
    x_new_processed = x_new_processed.astype('float32') / 255.0

    prediction = model.predict(x_new_processed)

    predicted_label = np.argmax(prediction, axis=1)[0]

    predicted_probability = prediction[0][predicted_label]
    
    predicted_index = data[data['label'] == predicted_label]['Vetement'].index[0]

    predicted_cloth = data.loc[predicted_index, 'Vetement']
    
    print("Predicted Label:", predicted_label)
    print("Predicted Probability:", predicted_probability)
    print("predicted cloth:", predicted_cloth)
```
